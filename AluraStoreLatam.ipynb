### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular la facturación total para cada tienda.
facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

# Mostramos al usuario la facturación total de cada tienda.
print(f'Facturación Tienda 1: {facturacion_tienda1}')
print(f'Facturación Tienda 2: {facturacion_tienda2}')
print(f'Facturación Tienda 3: {facturacion_tienda3}')
print(f'Facturación Tienda 4: {facturacion_tienda4}')
```

# 2. Ventas por categoría

```python
# Mostramos la categoríá de los productos.
categorias_tienda1 = tienda['Categoría del Producto'].value_counts()
categorias_tienda2 = tienda2['Categoría del Producto'].value_counts()
categorias_tienda3 = tienda3['Categoría del Producto'].value_counts()
categorias_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("Categorías más populares en la Tienda 1:")
print(categorias_tienda1)

print("\nCategorías más populares en la Tienda 2:")
print(categorias_tienda2)

print("\nCategorías más populares en la Tienda 3:")
print(categorias_tienda3)

print("\nCategorías más populares en la Tienda 4:")
print(categorias_tienda4)
```

```python
import matplotlib.pyplot as plt

# Creamos un diccionario para almacenar nuestro top 5 categorías de cada tienda.
top_categorias = {
    'Tienda 1': tienda['Categoría del Producto'].value_counts().head(5),
    'Tienda 2': tienda2['Categoría del Producto'].value_counts().head(5),
    'Tienda 3': tienda3['Categoría del Producto'].value_counts().head(5),
    'Tienda 4': tienda4['Categoría del Producto'].value_counts().head(5),
}

# Generamos un gráfico para cada tienda.
for nombre_tienda, categorias in top_categorias.items():
    plt.figure(figsize=(10, 6))
    categorias.plot(kind='bar', color='skyblue')
    plt.title(f'Top 5 Categorías más Populares en {nombre_tienda}', fontsize=16)
    plt.xlabel('Categoría', fontsize=12)
    plt.ylabel('Cantidad de Ventas', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Asignamos una variable de de promedio para cada tienda.
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

# Mostramos el la calificación promedio de cada tienda.
print(f'Calificación promedio de: Tienda 1 -> {promedio_tienda1:.2f}')
print(f'Calificación promedio de: Tienda 2 -> {promedio_tienda2:.2f}')
print(f'Calificación promedio de: Tienda 3 -> {promedio_tienda3:.2f}')
print(f'Calificación promedio de: Tienda 4 -> {promedio_tienda4:.2f}')
```

```python
import matplotlib.pyplot as plt

# Preparamos los datos para el gráfico.
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones_promedio = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]

# Creamos el gráfico de barras.
plt.figure(figsize=(10, 6))
plt.bar(nombres_tiendas, calificaciones_promedio, color='orange')
plt.title('Calificación Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Para cada tienda, obtenemos el conteo de ventas por producto.
productos_tienda1 = tienda['Producto'].value_counts()
productos_tienda2 = tienda2['Producto'].value_counts()
productos_tienda3 = tienda3['Producto'].value_counts()
productos_tienda4 = tienda4['Producto'].value_counts()
```

```python
# Mostramos los productos más vendidos de la Tienda 1.
print("Producto más vendido en la Tienda 1:")
print(productos_tienda1.head(1))

# Muestra los productos menos vendidos de la Tienda 1.
print("\nProducto menos vendido en la Tienda 1:")
print(productos_tienda1.tail(1))

# Mostramos los productos más vendidos de la Tienda 2.
print("Producto más vendido en la Tienda 2:")
print(productos_tienda2.head(1))

# Muestra los productos menos vendidos de la Tienda 2.
print("\nProducto menos vendido en la Tienda 2:")
print(productos_tienda2.tail(1))

# Mostramos los productos más vendidos de la Tienda 3.
print("Producto más vendido en la Tienda 3:")
print(productos_tienda3.head(1))

# Muestra los productos menos vendidos de la Tienda 3.
print("\nProducto menos vendido en la Tienda 3:")
print(productos_tienda3.tail(1))

# Mostramos los productos más vendidos de la Tienda 4.
print("Producto más vendido en la Tienda 4:")
print(productos_tienda4.head(1))

# Muestra los productos menos vendidos de la Tienda 4.
print("\nProducto menos vendido en la Tienda 4:")
print(productos_tienda4.tail(1))
```

```python
import matplotlib.pyplot as plt

# Creamos un diiccionario para almacenar los top 5 productos de cada tienda.
top_productos = {
    'Tienda 1': productos_tienda1.head(5),
    'Tienda 2': productos_tienda2.head(5),
    'Tienda 3': productos_tienda3.head(5),
    'Tienda 4': productos_tienda4.head(5),
}

# Generamos un gráfico para cada tienda.
for nombre_tienda, productos in top_productos.items():
    plt.figure(figsize=(10, 6))
    productos.plot(kind='bar', color='purple')
    plt.title(f'Top 5 Productos Más Vendidos en la {nombre_tienda}', fontsize=16)
    plt.xlabel('Producto', fontsize=12)
    plt.ylabel('Cantidad de Ventas', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
# Asignamos una variable del promedio del envío a cada tienda.
promedio_envio_tienda1 = tienda['Costo de envío'].mean()
promedio_envio_tienda2 = tienda2['Costo de envío'].mean()
promedio_envio_tienda3 = tienda3['Costo de envío'].mean()
promedio_envio_tienda4 = tienda4['Costo de envío'].mean()

# Mostramos el costo de envío al usuario.
print(f'Costo de envío promedio Tienda 1: {promedio_envio_tienda1:.2f}')
print(f'Costo de envío promedio Tienda 2: {promedio_envio_tienda2:.2f}')
print(f'Costo de envío promedio Tienda 3: {promedio_envio_tienda3:.2f}')
print(f'Costo de envío promedio Tienda 4: {promedio_envio_tienda4:.2f}')
```

```python
import matplotlib.pyplot as plt

# Preparamos los datos para el gráfico.
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_promedio = [promedio_envio_tienda1, promedio_envio_tienda2, promedio_envio_tienda3, promedio_envio_tienda4]

# Creamos el gráfico.
plt.figure(figsize=(10, 6))
plt.bar(nombres_tiendas, costos_promedio, color='teal')
plt.title('Costo de Envío Promedio por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Envío Promedio (en USD)', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

# **Análisis de Datos**

A continuación, se presentan los hallazgos detallados que respaldan la recomendación.

## **1. Facturación Total**

La Tienda 4 generó la menor facturación total, con un valor de 1,038,375,700.0. En contraste, la tienda con mayor facturación fue la Tienda 1, con 1,150,880,400.0. Este es el primer indicador de un rendimiento inferior.

## **2. Popularidad de Categorías**

El análisis de categorías reveló que, aunque las categorías más populares son similares en todas las tiendas, la Tienda 4 tiene una concentración de ventas en categorías como Artículos para el hogar y Libros, que si bien son importantes, no compensan el volumen de ventas general.

## **3. Calificación Promedio de Clientes**

La Tienda 4 registró la calificación promedio más baja de los clientes, con 4.00 estrellas. Aunque es una calificación buena, es inferior a la de las otras tiendas, lo que sugiere un nivel de satisfacción ligeramente más bajo que podría afectar su rendimiento a largo plazo. La Tienda 3 tiene la calificación promedio más alta con 4.05.

## **4. Rendimiento de Productos**

El producto más vendido en la Tienda 4 fue Cama box (62 unidades), mientras que el menos vendido fue Guitarra eléctrica (33 unidades). El rendimiento de sus productos estrella y de sus productos menos vendidos es comparable al de las otras tiendas, pero esto no se traduce en una facturación total más alta.

## **5. Costo de Envío Promedio**

El costo de envío promedio de la Tienda 4 fue el más bajo, con 23,459.46. A pesar de tener un costo logístico más bajo, su facturación sigue siendo la más baja, lo cual es una señal de que otros factores (como el bajo volumen de ventas) son más significativos.

# **Conclusión y Recomendación Final**

Con base en los datos presentados, se concluye que la Tienda 4 es la candidata más adecuada para ser vendida. A pesar de sus bajos costos de envío, su facturación total es la más baja de la cadena. Esto, combinado con una calificación de clientes ligeramente inferior a las demás, la sitúa como la opción menos rentable. La venta de esta tienda permitirá al Sr. Juan liberar recursos para invertir en las sucursales más exitosas.
